```python
x = df.iloc[:,1:-1]
```

```python
x = x.values
```

```python
plt.scatter(x[:,0],x[:,1],s = 10,alpha = 0.5)
```

```python
from sklearn.cluster import KMeans, MeanShift
```

```python
kmeans = KMeans(n_clusters = 3)           
```

```python
kmeans.fit(x)
```

```python
kmeans.predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],s = 10, c=kmeans.predict(x)) #c = colors
```

```python
df['Cluster_no'] = kmeans.predict(x)
```
